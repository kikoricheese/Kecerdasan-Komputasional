```python
# Blok 1
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Perceptron
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import numpy as np
from sklearn.ensemble import BaggingClassifier
```

- ***from sklearn import datasets*** : Mengimport modul datasets dari scikit-learn, yang menyediakan dataset yang sudah siap digunakan untuk eksperimen.
- ***from sklearn.preprocessing import StandardScaler*** : Mengimport kelas StandardScaler dari modul preprocessing yang digunakan untuk melakukan penskalaan fitur.
- ***from sklearn.linear_model import Perceptron*** : Mengimport kelas Perceptron dari modul linear_model yang digunakan sebagai base classifier.
- ***from sklearn.model_selection import train_test_split*** : Mengimport fungsi train_test_split dari modul model_selection yang digunakan untuk membagi dataset menjadi train set dan test set.
- ***from sklearn.metrics import accuracy_score*** : Mengimport fungsi accuracy_score dari modul metrics yang digunakan untuk mengukur akurasi prediksi.
import numpy as np: Mengimport modul numpy dengan alias np yang digunakan untuk manipulasi array.
- ***from sklearn.ensemble import BaggingClassifier*** : Mengimport kelas BaggingClassifier dari modul ensemble yang digunakan untuk melakukan ensemble learning dengan metode bagging.

```python
# Blok 2
wine = datasets.load_wine()
X_wine = wine.data
y_wine = wine.target
```

- ***datasets.load_wine()*** merupakan fungsi dari scikit-learn yang digunakan untuk memuat dataset Wine. Dataset Wine ini adalah salah satu dataset bawaan dari scikit-learn yang berisi data kualitas anggur dari tiga jenis kelas atau varietas anggur.
- Setelah memuat dataset Wine, atribut-atribut data disimpan dalam **X_wine** dan label kelas disimpan dalam **y_wine**.

- **y_wine** adalah variabel yang menyimpan label kelas dari dataset Wine.
Dengan menggunakan slicing [0:5], kita mengambil lima elemen pertama dari y_wine dan mencetaknya.

```python
# Blok 5
X_wine[:5]
```

- **X_wine** adalah variabel yang menyimpan atribut dari dataset Wine.
Dengan menggunakan slicing [:5], kita mengambil lima baris pertama dari X_wine dan mencetaknya.

```python
# Blok 6
X_train_wine, X_test_wine, y_train_wine, y_test_wine = train_test_split(X_wine, y_wine, test_size=0.2, random_state=0)
```

- ***X_wine dan y_wine adalah atribut dan target dari dataset Wine.
train_test_split()*** adalah fungsi yang digunakan untuk membagi dataset menjadi data latih dan data uji. Parameter X_wine dan y_wine merupakan input pertama yang akan dibagi. Parameter test_size=0.2 menentukan bahwa data uji akan berjumlah 20% dari keseluruhan dataset. Parameter random_state=0 digunakan untuk mengatur seed yang digunakan dalam proses pemilihan data uji secara acak, sehingga memastikan bahwa hasil yang sama akan diperoleh ketika kode dijalankan kembali dengan parameter yang sama.

- Setelah memanggil fungsi **train_test_split()**, data latih dan data uji akan dibagi menjadi X_train_wine, X_test_wine, y_train_wine, dan y_test_wine.

```python
# Blok 7
sc = StandardScaler()
X_train_wine_std = sc.fit_transform(X_train_wine)
X_test_wine_std = sc.transform(X_test_wine)
```

- ***StandardScaler()*** adalah objek dari kelas StandardScaler yang akan digunakan untuk melakukan standarisasi.
- ***sc.fit_transform(X_train_wine)*** digunakan untuk melakukan standarisasi pada data latih (X_train_wine). Metode fit_transform() digunakan untuk menghitung rata-rata dan standar deviasi dari setiap fitur pada data latih, dan kemudian mentransformasikan data latih tersebut dengan menggunakan rata-rata dan standar deviasi yang dihitung sebelumnya.
- ***sc.transform(X_test_wine)*** digunakan untuk melakukan transformasi pada data uji (X_test_wine) dengan menggunakan rata-rata dan standar deviasi yang telah dihitung dari data latih. Hal ini penting agar data uji memiliki skala yang sama dengan data latih, sehingga model yang akan dilatih dapat diterapkan secara konsisten pada data uji.
- Setelah proses standarisasi, hasilnya disimpan dalam variabel X_train_wine_std (data latih yang telah distandarisasi) dan X_test_wine_std (data uji yang telah distandarisasi).

```python
# Blok 8
test_sizes = [0.2, 0.3, 0.4]
for test_size in test_sizes:
    print(f"\nTest Size: {test_size}\n")
    X_train_wine, X_test_wine, y_train_wine, y_test_wine = train_test_split(X_wine, y_wine, test_size=test_size, random_state=0)
    X_train_wine_std = sc.fit_transform(X_train_wine)
    X_test_wine_std = sc.transform(X_test_wine)
    ppn_wine = Perceptron(eta0=0.1, random_state=0)
    ppn_wine.fit(X_train_wine_std, y_train_wine)
    y_pred_wine = ppn_wine.predict(X_test_wine_std)
    print('Accuracy on Wine dataset: %.2f' % accuracy_score(y_test_wine, y_pred_wine))
```

- ***test_sizes*** adalah daftar yang berisi beberapa ukuran test set yang akan digunakan dalam iterasi.
- ***for test_size in test_sizes*** : melakukan iterasi untuk setiap ukuran test set yang ada.
- Pada setiap iterasi, train_test_split digunakan untuk membagi dataset Wine (X_wine dan y_wine) menjadi data latih dan data uji, dengan ukuran test set ditentukan oleh test_size.
- Setelah membagi dataset, dilakukan standarisasi pada data latih dan data uji menggunakan objek sc yang telah didefinisikan sebelumnya.
Model Perceptron dengan parameter eta0=0.1 dan random_state=0 didefinisikan dengan nama ppn_wine.
- Model Perceptron dilatih pada data latih yang telah distandarisasi (X_train_wine_std dan y_train_wine) menggunakan metode fit().
- Selanjutnya, dilakukan prediksi pada data uji yang telah distandarisasi (X_test_wine_std) menggunakan metode predict(), dan hasil prediksi disimpan dalam variabel y_pred_wine.
- Akurasi dari model Perceptron dihitung dengan membandingkan hasil prediksi (y_pred_wine) dengan label sebenarnya (y_test_wine) menggunakan metode accuracy_score().
- Hasil akurasi dicetak ke layar dengan menggunakan print().

```python
# Blok 9
test_sizes = [0.2, 0.3, 0.4]
for test_size in test_sizes:
    print(f"\nTest Size: {test_size}\n")
    X_train_wine, X_test_wine, y_train_wine, y_test_wine = train_test_split(X_wine, y_wine, test_size=test_size, random_state=0)
    X_train_wine_std = sc.fit_transform(X_train_wine)
    X_test_wine_std = sc.transform(X_test_wine)
    ppn_wine = Perceptron(eta0=0.1, random_state=0)
    bc_wine = BaggingClassifier(base_estimator=ppn_wine, n_estimators=300, n_jobs=-1)
    bc_wine.fit(X_train_wine_std, y_train_wine)
    y_pred_wine = bc_wine.predict(X_test_wine_std)
    accuracy = accuracy_score(y_test_wine, y_pred_wine)
    print('Accuracy of Bagging Classifier on Wine dataset: {:.3f}'.format(accuracy))
```

- ***test_sizes*** adalah daftar yang berisi beberapa ukuran test set yang akan digunakan dalam iterasi.
- ***for test_size in test_sizes*** : melakukan iterasi untuk setiap ukuran test set yang ada.
- Pada setiap iterasi, train_test_split digunakan untuk membagi dataset Wine (X_wine dan y_wine) menjadi data latih dan data uji, dengan ukuran test set ditentukan oleh test_size.
- Setelah membagi dataset, dilakukan standarisasi pada data latih dan data uji menggunakan objek sc yang telah didefinisikan sebelumnya.
- Model Perceptron dengan parameter eta0=0.1 dan random_state=0 didefinisikan dengan nama ppn_wine.
- Model Bagging Classifier dengan base classifier Perceptron (ppn_wine) didefinisikan dengan nama bc_wine. Parameter n_estimators=300 menentukan jumlah estimator dalam ensemble, dan n_jobs=-1 mengizinkan penggunaan semua CPU yang tersedia untuk melatih model secara paralel.
- Model Bagging Classifier dilatih pada data latih yang telah distandarisasi (X_train_wine_std dan y_train_wine) menggunakan metode fit().
- Selanjutnya, dilakukan prediksi pada data uji yang telah distandarisasi (X_test_wine_std) menggunakan metode predict(), dan hasil prediksi disimpan dalam variabel y_pred_wine.
- Akurasi dari model Bagging Classifier dihitung dengan membandingkan hasil prediksi (y_pred_wine) dengan label sebenarnya (y_test_wine) menggunakan metode accuracy_score().
- Hasil akurasi dicetak ke layar dengan menggunakan print().

- Setelah model Bagging Classifier dilatih menggunakan data latih yang telah di-standarisasi (X_train_cancer_std dan y_train_cancer), prediksi dilakukan pada data uji yang telah di-standarisasi (X_test_cancer_std) menggunakan metode predict dari objek Bagging Classifier (bc_cancer), menghasilkan y_pred_cancer.

### 2. ulangi proses klasifikasi menggunakan bagging dengan base classifier perceptron pada dataset wine dan breastcancer. Gunakan perbandingan test set 0,3. Masing masing gunakan perbandingan n_estmators = 100, 200, dan 300. tuliskan akurasinya

## Klasifikasi pada Dataset Wine

```python
# Perbandingan test size
test_size = 0.3

# Perbandingan n_estimators
n_estimators = [100, 200, 300]

# Bagging Classifier dengan Perceptron pada Dataset Wine
print("Wine Dataset:")
for n in n_estimators:
    print(f"\nNumber of estimators: {n}")
    X_train_wine, X_test_wine, y_train_wine, y_test_wine = train_test_split(X_wine, y_wine, test_size=test_size, random_state=0)
    sc = StandardScaler()
    X_train_wine_std = sc.fit_transform(X_train_wine)
    X_test_wine_std = sc.transform(X_test_wine)
    ppn_wine = Perceptron(eta0=0.1, random_state=0)
    bc_wine = BaggingClassifier(base_estimator=ppn_wine, n_estimators=n, n_jobs=-1)
    bc_wine.fit(X_train_wine_std, y_train_wine)
    y_pred_wine = bc_wine.predict(X_test_wine_std)
    accuracy_wine = accuracy_score(y_test_wine, y_pred_wine)
    print('Accuracy on Wine dataset: {:.3f}'.format(accuracy_wine))
```

- Pertama, kita menginisialisasi test_size yang bernilai 0.3, yaitu 30% dari total data akan digunakan sebagai test set.
- Selanjutnya, kita mendefinisikan daftar n_estimators yang berisi nilai-nilai yang ingin dibandingkan, yaitu [100, 200, 300], yang mewakili jumlah estimator yang akan digunakan dalam Bagging Classifier.
- Kemudian, dilakukan perulangan for untuk setiap nilai n dalam n_estimators.
Pada setiap iterasi, kita mencetak nilai n untuk menunjukkan jumlah estimator yang digunakan.
- Selanjutnya, dataset Wine (X_wine dan y_wine) dibagi menjadi data latih (X_train_wine dan y_train_wine) dan data uji (X_test_wine dan y_test_wine) menggunakan fungsi train_test_split. Ukuran test set ditentukan oleh test_size.
Data latih dan data uji kemudian di-standarisasi menggunakan objek sc yang dibuat.
- Selanjutnya, sebuah objek Perceptron (ppn_wine) dibuat dengan menggunakan parameter eta0=0.1 dan random_state=0. Objek ini akan menjadi base classifier untuk Bagging Classifier.
- Objek Bagging Classifier (bc_wine) dibuat dengan menggunakan base classifier Perceptron (ppn_wine) dan jumlah estimator yang ditentukan (n). n_jobs=-1 digunakan untuk menggunakan semua CPU yang tersedia dalam pelatihan model.
- Setelah model Bagging Classifier dilatih menggunakan data latih yang telah di-standarisasi (X_train_wine_std dan y_train_wine), prediksi dilakukan pada data uji yang telah di-standarisasi (X_test_wine_std) menggunakan metode predict dari objek Bagging Classifier (bc_wine), menghasilkan y_pred_wine.
Akurasi klasifikasi pada dataset Wine dengan menggunakan Bagging Classifier dicetak menggunakan fungsi accuracy_score untuk membandingkan y_test_wine (label sebenarnya) dengan y_pred_wine (label yang diprediksi).
