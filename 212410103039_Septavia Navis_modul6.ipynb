###2. untuk dataset iris carilah subset kombinasi dari attribut yang memeberi hasil klasifikasi lebih baik dari data asli

```python
from sklearn.feature_selection import RFE
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.datasets import load_iris
import pandas as pd

# Load dataset iris
iris = load_iris()

# Membuat DataFrame dari dataset
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)

# Membuat target variable
target = iris.target

# Membuat model pembelajaran mesin (Random Forest Classifier)
model = RandomForestClassifier()

# Menerapkan Recursive Feature Elimination (RFE) dengan model yang telah ditentukan
rfe = RFE(estimator=model, n_features_to_select=2)
rfe.fit(iris_df, target)

# Menampilkan hasil subset kombinasi atribut terbaik
selected_features = iris_df.columns[rfe.support_]
print("Subset kombinasi atribut terbaik:")
for feature in selected_features:
    print(feature)

# Split data menjadi training dan testing
X_train, X_test, y_train, y_test = train_test_split(iris_df[selected_features], target, test_size=0.4, random_state=0)

# Membuat dictionary untuk menyimpan akurasi dari setiap kombinasi atribut
attribute_combination_scores = {}

# Melakukan klasifikasi menggunakan Gaussian Naive Bayes untuk setiap kombinasi atribut dan menyimpan skor akurasinya
for attributes in selected_features:
    model = GaussianNB()
    model.fit(X_train[[attributes]], y_train)
    score = model.score(X_test[[attributes]], y_test)
    attribute_combination_scores[str([attributes])] = score

print(' ')
# Menampilkan akurasi dari setiap kombinasi atribut yang telah dicoba
for attributes, score in attribute_combination_scores.items():
    print("Subset kombinasi atribut terbaik:", attributes)
    print("Akurasi:", score)
    print()
```

Dalam kode di atas, pertama-tama kita menerapkan Recursive Feature Elimination (RFE) menggunakan Random Forest Classifier untuk memilih subset kombinasi atribut terbaik dari dataset iris. Selanjutnya, kita membagi data menjadi training dan testing menggunakan train_test_split. Kemudian, kita melakukan klasifikasi menggunakan Gaussian Naive Bayes untuk setiap kombinasi atribut yang telah ditentukan, dan menyimpan skor akurasinya dalam dictionary attribute_combination_scores. Terakhir, kita mencetak hasil subset kombinasi atribut terbaik dan akurasinya menggunakan loop for.
