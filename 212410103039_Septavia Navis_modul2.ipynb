2. Jelaskan proses penyusunan dataset!

  **Jawab:**

  Proses penyusunan dataset kita memerlukan sebuah ekspresi atau formula yang dapat mengecek satu persatu data yang memenuhi formula subset yang telah dibentuk. Sebagai contoh, kita ingin memperoleh dataset tanpa nilai missing value pada variabel ***namavariabel***. 
  Untuk tahap penyusunan pemrosesan awal data adalah sebagai berikut:

  - **Agregasi** => Penggabungan obyek ke dalam sebuah obyek tunggal. Rumusnya : Sum, average, min, max
  - **Sampling** => Pemilihan bagian obyek data yang akan di analisis. Sample harus representatif (mewakili seluruh data) jika mempunyai sifat yang sama dengan seluruh data biasa diukur dengan rata-rata/mean. Penggunaan sample yang baik tidak menjamin bahwa hasil pemrosesan ***datamining*** pada sample sama bagusnya dengan pemrosesan pada seluruh data asli. Pendekatan sampling yaitu ***simple random sampling*** yang terdiri dari tanpa pengembalian dan dengan pengembalian.
  - **Binerisasi** => Transformasi data dari tipe continue, diskret menjadi tipe biner. Algoritma asosiasi membutuhkan data dengan atribut bertipe biner. Jumlah atribut yang dibutuhkan untuk binerisasi adalah **N=log2(M), M=jumlah kelas kategori**. Contoh: {rusak,jelek,sedang,bagus,sempurna}, M=5.
  - **Diskretisasi** => Transformasi data dari tipe kontinue ke diskrit.
  - **Pengurangan Dimensi** => Mengurangi jumlah waktu dan memory yang dibutuhhkan. Membuat data lebih mudah divisualisasi. Membantu mengurangi fitur-fitur yang tidak relevan/mengurangi gangguan. Teknik yang digunakan diantaranya, ***Principal Component Analysis (PCA)*** dan ***Singular Value Decomposition (SVD)***
  - **Feature Subset Selection** => Proses pencarian terhadap semua kemungkinan subset fitur dengan menghilangkan fitur yang redundan dan menghilangkan fitur yang tidak mengandung informasi yang berguna untuk pekerjaan. Teknik yang digunakan diantaranya: 
    1. ***Brute-force***, pada proses data mining dilakukan dengan mencoba semua fitur.
    2. ***Filtering***, memilih fitur sebelum proses data mining dilakukan.
    3. ***Wrapper***, menggunakan algoritma data mining untuk memilik subset fitur yang paling baik.
  
    Proses pemilihan fitur adalah sebagai berikut:
    
      1. Melakukan pengukuran untuk evaluasi suatu subset fitur
      2. Menggunakan metode pencarian mengontrol pemilihan subset fitur baru
      3. Menggunakan kriteria untuk melakukan penghentian proses
      4. Menggunakan validasi
    
  - **Feature Transformation** => Proses memetakan keseluruhan himpunan nilai dan fitur-fitur yang diberikan ke suatu subset nilai pengganti. Metode yang digunakan adalah **Standarisasi (median, standar deviasi)** dan **Normalization, dimana data sebuah atribut diskalakan ke dalam rentang yang ditentukan.

8. Buatlah pararrel coordinat plot dari data yang berkelas setosa, versicolor, dan virginica. Berikan warna merah untuk setosa, hijau untuk versicolor, biru untuk virginica. Hint gunakan perintah plot dan operasi slicing pada numpy untuk mempermudah tugas anda!

  **Jawab:**

```python
x = np.array(['Petal Width', 'Petal Length', 'Sepal Width', 'Sepal Length'])

for i in range(len(setosa)):
    plt.plot(x, setosa[i,:], color='red')
for i in range(len(verginica)):
    plt.plot(x, verginica[i,:], color='blue')
for i in range(len(versicolor)):
    plt.plot(x, versicolor[i,:], color='green')

plt.show()
```

untuk membuat plot menggunakan line chart dengan sumbu x berisi label fitur (Petal Width, Petal Length, Sepal Width, dan Sepal Length) dan sumbu y berisi nilai setiap fitur pada setiap spesies iris.

Karena tujuan dari plot ini adalah untuk membandingkan setiap spesies iris, maka setiap plot untuk setiap spesies iris diberikan warna yang berbeda. Plot untuk setiap spesies iris diwarnai dengan warna merah (red) untuk setosa, biru (blue) untuk virginica, dan hijau (green) untuk versicolor.
