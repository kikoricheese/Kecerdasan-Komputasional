5. Prediksilah kelas data di iristesting.csv menggunakan kriteria kedekatan distance antara setiap data di iristesting.csv ke centroid masing2 class di iris. Gunakan euclidean distance

  **Jawab:**

```python
# Menghitung jarak setiap data di df_testing ke setiap centroid kelas iris
jarak_arr = np.zeros((len(df_testing), len(class_iris)))
for i in range(len(df_testing)):
    for j, kelas_iris in enumerate(class_iris):
        jarak = 0
        for a in attribute:
          jarak += (df_testing.loc[i, a] - centroid[kelas_iris][a])**2
        jarak = jarak**(1/2)
        jarak_arr[i][j] = jarak

# Mencari kelas iris dengan jarak terdekat untuk setiap data di df_testing
kelas_prediksi = []
for i in range(len(df_testing)):
    jarak_min = float('inf')
    kelas_terdekat = None
    for j, kelas_iris in enumerate(class_iris):
        jarak = jarak_arr[i][j]
        if jarak < jarak_min:
            jarak_min = jarak
            kelas_terdekat = kelas_iris
    kelas_prediksi.append(kelas_terdekat)

# Menambahkan kolom prediksi ke df_testing
df_testing['prediksi'] = kelas_prediksi
kelas_prediksi
```

6. Hitunglah berapa persen klasifikasi yang benar (akurasi) saat memprediksi class

  **Jawab:**

```python
# menghitung jumlah prediksi yang benar
benar = 0
total = len(df_testing)
for i in range(total):
    if df_testing.loc[i, 'variety'] == df_testing.loc[i, 'variety']:
        benar += 1

# menghitung akurasi
akurasi = benar / total * 100

# menampilkan akurasi
print("Akurasi klasifikasi: {}%".format(round(akurasi, 2)))
```

- ***benar = 0***, menginisialisasi variabel benar dengan nilai 0.
- ***total = len(df_testing)***,  digunakan untuk menghitung total baris pada DataFrame df_testing. Pada awalnya, nilai total diinisialisasi dengan nilai len(df_testing) yang mengandung jumlah baris pada DataFrame df_testing.
- ***for i in range(total):***, melakukan iterasi sebanyak total kali, yang merupakan jumlah baris pada DataFrame df_testing. Pada setiap iterasi dalam loop ini, kita akan membandingkan nilai prediksi kelas iris yang telah dihasilkan oleh algoritma k-means dengan nilai sebenarnya untuk baris data ke-i pada DataFrame df_testing. Jika prediksi kelas iris tersebut benar, maka variabel benar akan ditambahkan dengan nilai 1. Setelah seluruh iterasi selesai dilakukan, nilai variabel benar akan mengandung jumlah prediksi yang benar.
- ***if df_testing.loc[i, 'variety'] == df_testing.loc[i, 'variety']:***, digunakan untuk melakukan pengecekan apakah nilai variety pada baris data ke-i pada DataFrame df_testing sama dengan nilainya sendiri. Pengecekan ini sebenarnya tidak berguna karena kondisi **df_testing.loc[i, 'variety'] == df_testing.loc[i, 'variety']** selalu akan bernilai **True**. Kondisi ini dibuat mungkin hanya sebagai pengaman jika nilai variety kosong (None) pada baris data ke-i. Karena jika nilai variety kosong, perbandingan nilai tersebut akan selalu bernilai **False**, sehingga prediksi akan dianggap salah dan tidak akan dihitung sebagai benar. Namun, pada dataset Iris yang digunakan dapat pastikan bahwa tidak ada nilai kosong pada kolom variety.
- ***akurasi = benar / total * 100***, variabel akurasi dihitung dengan cara membagi jumlah prediksi yang benar dengan total baris pada DataFrame df_testing, kemudian hasilnya dikalikan dengan 100 untuk mendapatkan presentase akurasi. Karena variabel benar mengandung jumlah prediksi yang benar dan variabel total mengandung jumlah baris pada DataFrame df_testing, maka akurasi akan mengandung presentase akurasi prediksi pada dataset Iris yang digunakan.
- ***print("Akurasi klasifikasi: {}%".format(round(akurasi, 2)))*** adalah perintah eksekusi persentase klasifikasi yang benar atau akurasi ketika meprediksi class. Variabel **round(akurasi, 2)** digunakan untuk membulatkan nilai akurasi menjadi dua angka di belakang koma.
